```python
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
train_path = "/kaggle/input/regression-model-dataset/train (2).csv"
train_data = pd.read_csv(train_path)
```

```python
test_path = "/kaggle/input/regression-model-dataset/test (1).csv"
test_data = pd.read_csv(test_path)
```

```python
data = pd.read_csv(train_path)
```

```python
data.head()
```

```python
train_X = train_data.drop("SalePrice", axis=1)
train_y = train_data.loc[:, "SalePrice"]
```

```python
test_X = test_data.drop("SalePrice", axis=1)
test_y = test_data.SalePrice
```

```python
linear_reg = LinearRegression()
```

```python
ridge_reg = Ridge(alpha=0.05)
```

```python
linear_reg.fit(train_X,train_y)
```

```python
ridge_reg.fit(train_X, train_y)
```

```python
linear_pred = linear_reg.predict(test_X)
ridge_pred = ridge_reg.predict(test_X)
```

```python
linear_mse = mean_squared_error(test_y, linear_pred)
ridge_mse = mean_squared_error(test_y, ridge_pred)
```

```python
print(f"MSE without Ridge: {linear_mse}")
print(f"MSE with Ridge: {ridge_mse}")  # ridge pred is better
```
